```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('openmp_new.csv')
```

```python
df24 = df[df["Threads"] == 24]

plt.figure(figsize=(10, 6))
plt.plot(df24["Events"], df24["OMP_Events"], marker='o', label="OMP Events only")
plt.plot(df24["Events"], df24["OMP_Events_and_k"], marker='s', label="OMP Events and $k$")
plt.plot(df24["Events"], df24["OMP_SIMD"], marker='^', label="OMP Events and SIMD $k$")

plt.xscale("log")
plt.xlabel("Events")
plt.ylabel("Speedup relative to serial")
plt.title("Comparison of OMP Speedup Methods across Number of Events (24 Threads)")
plt.legend()
plt.grid(True, which="both", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plot = True
plt.figure(figsize=(10, 6))
for events, group in df.groupby("Events"):
    group_sorted = group.sort_values("Threads")
    if events == 1000000:
        plt.plot(group_sorted["Threads"], group_sorted["OMP_Events"], marker='o', color='black', label=f"{events} events")
    elif plot:
        plt.plot(group_sorted["Threads"], group_sorted["OMP_Events"], marker='*', alpha=0.5, label=f"{events} events")
    plot = not plot


plt.xlabel("Thread count")
plt.ylabel("Speedup relative to serial")
plt.title("OMP \"Events Only\" Speedup by Thread for Different Event Counts")
plt.legend(title="Events")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
